```python
df = df.reset_index(drop=True)
tokens_length = df['code_tokens'].str.len()
df.insert(loc=2, column='tokens_length', value=tokens_length)
```

```python
df.sort_values(by='tokens_length', inplace=True)
```

```python
from pygments import highlight
from pygments.lexers import JavaLexer
from pygments.formatters import TerminalFormatter
```

```python
highlighted_code = highlight(df.loc[24, 'code'], JavaLexer(), TerminalFormatter())
print(highlighted_code)
```

```python
df = df[(50 < df['tokens_length']) & (df['tokens_length'] < 100)]
for index, snippet in df['code'].items():
    print('[INDEX] ' ,index)
    highlighted_code = highlight(snippet, JavaLexer(), TerminalFormatter())
    print(highlighted_code)
    print('==============================================')
```

```python
annotated_code = annotated_code.dropna()

def printEachSnippet(row):
    print(row['cwe'], '-', row['snippet_id'])
    highlighted_code = highlight(row['vulnerable_code'], JavaLexer(), TerminalFormatter())
    print(highlighted_code)
    print('---')

    highlighted_code = highlight(row['fixed_code'], JavaLexer(), TerminalFormatter())
    print(highlighted_code)
    print('**************************************************')
    


operation = annotated_code.apply(printEachSnippet, axis=1)
```
