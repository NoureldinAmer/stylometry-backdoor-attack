```python
import pandas as pd
import os
import subprocess
import json
import re
# from sarif_om import SarifLog, from_dict
```

```python
# Read the CSV file
df = pd.read_csv('vulnerability_results_anthropic.csv', index_col='Unnamed: 0')
```

```python
df
```

```python
# Create directories for code snippets
os.makedirs('snippets/original', exist_ok=True)
os.makedirs('snippets/modified', exist_ok=True)

# Iterate over the DataFrame and write code snippets to files
for idx, row in df.iterrows():
    orig_code = row['original_snippet']
    mod_code = row['modified_code']
    
    # Define file paths using the index as the filename
    orig_file_path = os.path.join("snippets", "original", f"{idx}.java")
    mod_file_path = os.path.join("snippets", "modified", f"{idx}.java")
    
    # Write the original and modified code to respective files
    with open(orig_file_path, 'w') as f:
        f.write(orig_code)
    with open(mod_file_path, 'w') as f:
        f.write(mod_code)
```
